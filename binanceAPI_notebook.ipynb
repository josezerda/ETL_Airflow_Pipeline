# Metodo A consulta a Binance

```python
import json 
import requests 
  
# defining key/request url 
key = "https://api.binance.com/api/v3/ticker/price?symbol=BTCUSDT"
  
# requesting data from url 
data = requests.get(key)   
data = data.json() 
print(f"{data['symbol']} price is {data['price']}") 
```

```python

# Import libraries 
import json 
import requests 
  
# Defining Binance API URL 
key = "https://api.binance.com/api/v3/ticker/price?symbol="
  
# Making list for multiple crypto's 
currencies = ["BTCUSDT", "ETHUSDT", "BNBUSDT"] 
j = 0
  
# running loop to print all crypto prices 
for i in currencies: 
    
    # completing API for request 
    url = key+currencies[j]   
    data = requests.get(url) 
    data = data.json()
    print(data) 
    j = j+1
    print(f"{data['symbol']} price is {data['price']}") 
```

# Metodo B consulta a Binance

```python
from websocket import WebSocketApp
import json

def on_message(ws, msg):
    print(msg)
    tick = json.loads(msg)
    if tick['k']['x']:
        print("NUEVA VEL")
    print(tick['k']['c'])


ws = WebSocketApp("wss://stream.binance.com:9443/ws/btcusdt@kline_1m", on_message=on_message)
ws.run_forever()
```

# Metodo C, consulta valores NYSE

c *number
The close price for the symbol in the given time period.

h *number
The highest price for the symbol in the given time period.

l *number
The lowest price for the symbol in the given time period.

n integer
The number of transactions in the aggregate window.

o *number
The open price for the symbol in the given time period.

o tcboolean
Whether or not this aggregate is for an OTC ticker. This field will be left off if false.

t *integer
The Unix Msec timestamp for the start of the aggregate window.

v *number
The trading volume of the symbol in the given time period.

v wnumber
The volume weighted average price.

```python
import yfinance as yf
import datetime

# Especificar el símbolo de la acción (por ejemplo, 'AAPL' para Apple)
symbol = "AAPL"  # Reemplaza con el símbolo de la acción que desees

# Establecer el rango de fechas (últimos 365 días)
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=365)

# Descargar los datos de la acción
data = yf.download(symbol, start=start_date, end=end_date)

# Mostrar las primeras filas del DataFrame
print(data.head())

# Guardar los datos en un archivo CSV
data.to_csv(f'{symbol}_historical_data.csv')
```
